<B>Ejercicio N°1:</B>
<p>
Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.

```python
import pandas as pd

# Calcula los ingresos por ventas
def calcular_ventas(num_unidades, precio_unitario):
    return num_unidades * precio_unitario

# Calcula las ventas con el descuento aplicado
def aplicar_descuento(ventas):
    return ventas * 0.95

# Diccionario para almacenar las ventas de cada día
ventas = {}
# Solicita al usuario el precio de venta unitario
precio_unitario = float(input('Ingrese el precio de venta unitario: '))

# Bucle para solicitar al usuario el número de unidades vendidas para cada día y calcular las ventas
for i in range(1, 6):
    num_unidades = float(input('Introduce el número de unidades vendidas el día ' + str(i) +': '))
    ventas[i] = calcular_ventas(num_unidades, precio_unitario)

# Crea una serie a partir del diccionario de ventas
ventas_serie = pd.Series(ventas)
# Muestra en pantalla las ventas del día antes del descuento
print('\nVentas del día (sin descuento):\n', ventas_serie)
# Muestra en pantalla las ventas del día con descuento
print('\nVentas del día (con descuento del 5%):\n', aplicar_descuento(ventas_serie))
```

<B>Ejercicio N°2:</B><p>
Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.
El resultado final debe ser una serie con los estadísticos.

```python
import pandas as pd

def estadisticos_ingresos_ventas(ventas):
    # Convertir el diccionario de ventas en una serie
    ventas_serie = pd.Series(ventas)

    # Calcular el ingreso mínimo, máximo y medio directamente desde la serie
    estadisticos = pd.Series({
        'Ingreso Mínimo': ventas_serie.min(),
        'Ingreso Máximo': ventas_serie.max(),
        'Ingreso Medio': ventas_serie.mean()
    })

    return estadisticos

# Diccionario de ventas mensuales de los clientes
ventas = {'Cliente_01': 1500, 'Cliente_02': 3200, 'Cliente_03': 2800, 'Cliente_04': 4100, 'Cliente_05': 3600}

# Muestra en pantalla los estadísticos de las ventas
print(estadisticos_ingresos_ventas(ventas))
```

Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la tabla siguiente:<p>

<table border="1" >
  <tr>

    <th>Ventas</th>  
    <th>Gastos</th>
  </tr>
  <tr>
    <td>Enero</td>
    <td>30500</td>
    <td>22000</td>
  </tr>
  <tr>
    <td>Febrero</td>
    <td>35600</td>
    <td>23400</td>
  </tr>
  <tr>
    <td>Marzo</td>
    <td>28300</td>

    <td>Abril</td>
    <td>33900</td>

  </tr>

```python
import pandas as pd

# Crea un diccionario con los datos
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Ventas': [30500, 35600, 28300, 33900],
    'Gastos': [22000, 23400, 18100, 20700]
}

# Crea el DataFrame
df = pd.DataFrame(datos)

# Muestra el DataFrame en pantalla
print(df)
```

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
# Agrega la columna 1: Gasto/Ventas*100
df['Gasto/Ventas*100'] = (df['Gastos'] / df['Ventas']) * 100

# Agrega la columna 2: Meta Superada o no
#df['Meta'] = df['Ventas'].apply(lambda x: 'Meta Superada' if x >= 30000 else 'Meta no Superada')
df['Meta'] = ['Meta Superada' if ventas >= 30000 else 'Meta no Superada' for ventas in df['Ventas']]

# Agrega la columna 3: Sobre gasto
#df['Sobre Gasto'] = df.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else 0, axis=1)
df['Sobre Gasto'] = [gastos - ventas if gastos > ventas else 0 for gastos, ventas in zip(df['Gastos'], df['Ventas'])]

# Obtiene el total de las ventas de los últimos 3 meses
total_ventas_ultimos_3_meses = df['Ventas'].tail(3).sum()

# Obtiene el total de gastos de febrero a abril
#total_gastos_febrero_abril = df.loc[df['Mes'].isin(['Febrero', 'Marzo', 'Abril']), 'Gastos'].sum()
total_gastos_febrero_abril = df.loc[1:3, 'Gastos'].sum()

# Muestra el DataFrame con las nuevas columnas
print(df)

# Muestra el total de ventas de los últimos 3 meses
print("Total de ventas de los últimos 3 meses:", total_ventas_ultimos_3_meses)

# Muestra el total de gastos de febrero a abril
print("Total de gastos de febrero a abril:", total_gastos_febrero_abril)
```
